# Log-Normal Distribution

It is a heavy tailed continuous probability distribution of a random variable whose logarithm is normally distributed.



image.png

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters of the log-normal distribution
mu = 2  # Mean of the logarithm of the distribution
sigma = 1  # Standard deviation of the logarithm of the distribution
```

```python
# Generate a range of x values
x = np.linspace(0.1, 150, 1000)

# Calculate the probability density function (PDF) of the log-normal distribution
pdf = (1 / (x * sigma * np.sqrt(2 * np.pi))) * np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2))
```

```python
# Plot the log-normal distribution
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, color='orange')
plt.title('Log-Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

### Characteristics:

* The log-normal distribution is defined only for positive real numbers.
* It is characterized by two parameters:

   - the mean () and the standard deviation () of the associated normal distribution of the natural logarithm.

### Real Life Examples of Log-Normal Distribution:

1. The length of comments posted on your favorite YouTuber's channel follows a log-normal distribution, with most comments being short and concise, while fewer comments are longer and more detailed, resembling a long right-tailed distribution.

2. Users interacting with news apps and online articles exhibit a log-normal distribution of engagement metrics, with a concentration of users at lower engagement levels and a long right tail representing highly engaged users.

3. Log-normal distributions are often used to model income distributions in societies with significant income inequality (i.e long right tail of distribution captures the presence of high-income earners, while the bulk of the distribution represents lower and middle-income households).
